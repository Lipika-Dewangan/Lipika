```python
import aspect_based_sentiment_analysis as absa

nlp = absa.load()
text = ("Did not enjoy the new Windows 8 and touchscreen functions .")

asp1, asp2 = nlp(text, aspects=['Windows 8', 'touchscreen'])
print(asp1.sentiment)
print(asp2.sentiment)
```

```python
text_2 = ("BBSR is a wonderful city to live but it is costly place.")

asp3, asp4 = nlp(text_2, aspects=['BBSR', 'place'])
print(asp3.sentiment)
print(asp4.sentiment)
```

```python
recognizer = absa.aux_models.BasicPatternRecognizer()
nlp = absa.load(pattern_recognizer=recognizer)
completed_task = nlp(text=text_2, aspects=['BBSR', 'place'])
city, plc = completed_task.examples

absa.summary(city)
absa.display(city.review)

absa.summary(plc)
absa.display(plc.review)
```
